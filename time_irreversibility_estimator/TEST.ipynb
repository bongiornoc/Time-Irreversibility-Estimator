```python
import numpy as np

# Definiamo una funzione che prende un array T x M e restituisce un vettore di dimensione K
def my_function(matrix):
    sum0 = np.sum(matrix, axis=0)  # somma lungo l'asse 0
    sum1 = np.sum(matrix, axis=1)  # somma lungo l'asse 1
    return np.concatenate((sum0, sum1))  # restituisce la concatenazione

# Creiamo un array di dimensione N x T x M
N, T, M = 5, 3, 4
array = np.random.rand(N, T, M)

# Funzione wrapper per applicare my_function
def apply_my_function(arr):
    return my_function(arr)

# Vectorize la funzione
vectorized_function = np.vectorize(apply_my_function, signature='(m,n)->(k)')

# Applichiamo la funzione vettorializzata all'array
result = vectorized_function(array)

# Assicuriamoci che il risultato abbia la forma corretta N x K
K = result.shape[-1]
result = result.reshape(N, K)

print(result)
```
